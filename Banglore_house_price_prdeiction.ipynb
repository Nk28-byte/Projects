```python
def convertRange(x):
  if isinstance(x, str):  # Check if x is a string
    temp = x.split('-')
    if len(temp) == 2:
      return (float(temp[0]) + float(temp[1])) / 2
  try:
    return float(x)  # Convert to float if possible
  except (TypeError, ValueError):  # Handle TypeError and ValueError
    return None
```

```python
df['total_sqft']=df['total_sqft'].apply(convertRange)
```

# Outlier Detection And Removal

```python
df.describe()
```

```python
def remove_outliers_sqft(df):
  df_output=pd.DataFrame()
  for key,subdf in df.groupby('location'):
    m=np.mean(subdf.price_per_sqft)

    st=np.std(subdf.price_per_sqft)

    gen_df=subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <=(m+st))]
    df_output=pd.concat([df_output,gen_df],ignore_index=True)
  return df_output
df=remove_outliers_sqft(df)
df.describe()
```

```python
def bhk_outlier_remover(df):
  exclude_indices=np.array([])
  for location,location_df in df.groupby('location'):
    bhk_stats={}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk]={
          'mean':np.mean(bhk_df.price_per_sqft),
          'std':np.std(bhk_df.price_per_sqft),
          'count':bhk_df.shape[0]
      }
      for bhk, bhk_df in location_df.groupby('bhk'):
        stats=bhk_stats.get(bhk-1)
        if stats and stats['count']>5:
          exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python

df=bhk_outlier_remover(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape,x_test.shape
```

# Applying Linear Regression

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),  # Wrap in a tuple
    remainder='passthrough'
)
```

```python
scaler=StandardScaler()
```

```python

lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python

y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

# Applying Lasso

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

# Applying Ridge

```python
ridge=Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
print("No Regularisation:",r2_score(y_test,y_pred))
print("Lasso:",r2_score(y_test,y_pred))
print("Ridge:",r2_score(y_test,y_pred))
```
